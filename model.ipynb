```
import pandas as pd
import numpy as np
import re
```

```
tweets = pd.read_csv('tweets_SAlabel.csv')
```

```
from sklearn.model_selection import train_test_split, GridSearchCV
input = tweets['text']
target = tweets['label']
trainInput, testInput, trainTarget, testTarget = train_test_split(
    input, target, test_size=0.1, random_state=200)
```

```
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from time import time
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```
pipe = Pipeline([('vect', CountVectorizer()),
                     ('lr', LogisticRegression())])
tuned_parameters = {
    'vect__ngram_range': [(1, 1), (1, 2), (1, 3)],
}
grid = GridSearchCV(pipe, tuned_parameters, cv=10)
grid.fit(trainInput, trainTarget)
prediction = grid.predict(testInput)

print(classification_report(testTarget,prediction, digits=4))
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
print("Best estimator: ", grid.best_estimator_)
```

```
pipe = Pipeline([('vect', TfidfVectorizer()),
                     ('lr', LogisticRegression())])
tuned_parameters = {
    'vect__ngram_range': [(1, 1), (1, 2), (1, 3)],
    'vect__use_idf': (True, False),
    'vect__norm': ('l1', 'l2'),
}
grid = GridSearchCV(pipe, tuned_parameters, cv=10)
grid.fit(trainInput, trainTarget)
prediction = grid.predict(testInput)

print(classification_report(testTarget,prediction, digits=4))
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
print("Best estimator: ", grid.best_estimator_)
```
