```
from IPython import display
import math
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('/content/tweets_total_cleaned_v2.csv', usecols=[2,3,4])
df.info()
```

```
df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)
df.info()
```

```
total_len = df.shape[0]
total_len
```

```
df.iloc[:10]
```

```
df['neg_score'] = 0.0
df['neu_score'] = 0.0
df['pos_score'] = 0.0
df['compound'] = 0.0
df.iloc[:10]
```

```
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
 
sia = SIA()
# pol_score = sia.polarity_scores(df.at[1,'text'])
# print(pol_score)
# print(pol_score['neu'])
# df.at[1,'neg_score'] = pol_score['neg']
# df.at[1,'neu_score'] = pol_score['neu']
# df.at[1,'pos_score'] = pol_score['pos']
# df.at[1,'compound'] = pol_score['compound']
# df.at[0,'neu_score'] = 0.833
# df.iloc[:2]
```

```
for i in range(total_len):
  pol_score = sia.polarity_scores(df.at[i,'text'])
  df.at[i,'neg_score'] = pol_score['neg']
  df.at[i,'neu_score'] = pol_score['neu']
  df.at[i,'pos_score'] = pol_score['pos']
  df.at[i,'compound'] = pol_score['compound']
  print('Processing '+str(i+1)+'/'+str(total_len)+'##################################################')
```

```
df.iloc[:10]
```

```
df['label'] = 0
df.loc[df['compound']>0.2,'label'] = 1
df.loc[df['compound']<-0.2,'label'] = -1
df.iloc[:10]
```
